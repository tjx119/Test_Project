```python
import random
import itertools
from copy import deepcopy

def make_board(m = 3):
    
    numbers = list(range(1, m ** 2 + 1))
#board是个二维数组
    board = None
    while board == None:
        board = attempt_board(m, numbers)
    return board

def attempt_board(m, numbers):
    n = m ** 2
    board = [[None for _ in range(9)] for _ in range(9)]
    
    for i, j in list(itertools.product(range(9),repeat=2)):
        
        i0, j0 = i - i % m, j - j % m
        random.shuffle(numbers)
        for x in numbers:
            
            if (x not in board[i]        #判断竖排不同             
                and all(row[j] != x for row in board)   #判断行不同 
                and all(x not in row[j0:j0+m] for row in board[i0:i])): #判断9宫格不同
                
                board[i][j] =  x
                break
        else:
            return None                
    return board

#打印我们的数独列表
def print_board(board,m = 3):
    numbers = list(range(1, m**2 + 1))

    #每一行随机把5个数字变成None
    omit = 6   #omit变量掌控着每一行被抹去的数字个数
    challange = deepcopy(board)
    for i, j in itertools.product(range(omit), range(m ** 2)):
        x = random.choice(numbers) - 1
        challange[x][j] = None

    #打印出整个数独列表
    spacer = "++-----+-----+-----++-----+-----+-----++-----+-----+-----++"
    print (spacer.replace('-','='))
    for i, line in enumerate(challange):
        print("||  {}  |  {}  |  {}  ||  {}  |  {}  |  {}  ||  {}  |  {}  |  {}  ||"
              .format(*(cell or ' ' for cell in line)))
        if (i + 1) % 3 == 0: print(spacer.replace('-', '='))
        else: print(spacer)
    return challange

def print_answers(board):
    spacer = "++-----+-----+-----++-----+-----+-----++-----+-----+-----++"
    print (spacer.replace('-','='))
    for i, line in enumerate(board):
        print("||  {}  |  {}  |  {}  ||  {}  |  {}  |  {}  ||  {}  |  {}  |  {}  ||"
              .format(*(cell or ' ' for cell in line)))
        if (i + 1) % 3 == 0: print(spacer.replace('-', '='))
        else: print(spacer)
            
if __name__ == '__main__':
    Board = make_board(m=3)
    print_board(Board)
#    print_answers(Board)
```
